# Part - B Data Visualization

#### Importing libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
company_sales = pd.read_csv("company-sales.csv")
```

```python
company_sales.head()
```

### 1. Read Total profit of all months and show it using a line plot 

```python
plt.plot(company_sales['month_number'],company_sales['total_profit'], label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title("Total Profit of all months")
plt.legend()
plt.show()
```

### 2. Read all product sales data and show it using a multiline plot

```python
plt.figure(figsize=(16,7))
plt.plot(company_sales['facecream'],label='Face Cream Sales Data')
plt.plot(company_sales['facewash'],label='Face Wash Sales Data')
plt.plot(company_sales['toothpaste'],label='Toothpaste Sales Data')
plt.plot(company_sales['bathingsoap'],label='Bathing Soap Sales Data')
plt.plot(company_sales['shampoo'],label='Shampoo Sales Data')
plt.plot(company_sales['moisturizer'],label='Moisturizer Sales Data')
plt.ylabel('Sales Data')
plt.title('Product Sales Data')
plt.legend()
plt.show()
```

### 3 . Display the number of units sold per month for each product using multiline plots. (i.e.,Separate Plotline for each product).

```python
plt.figure(figsize=(16,7))
plt.plot(company_sales['month_number'],company_sales['facecream'],label='Face Cream Sales Data')
plt.plot(company_sales['month_number'],company_sales['facewash'],label='Face Wash Sales Data')
plt.plot(company_sales['month_number'],company_sales['toothpaste'],label='Toothpaste Sales Data')
plt.plot(company_sales['month_number'],company_sales['bathingsoap'],label='Bathing Soap Sales Data')
plt.plot(company_sales['month_number'],company_sales['shampoo'],label='Shampoo Sales Data')
plt.plot(company_sales['month_number'],company_sales['moisturizer'],label='Moisturizer Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Data')
plt.title('Product Sales Data')
plt.legend()
plt.show()
```

### 4. Read toothpaste sales data of each month and show it using a scatter plot.

```python
plt.scatter(company_sales['month_number'],company_sales['toothpaste'], label='Toothpaste Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales Data')
plt.title('Toothpaste Sales Data of each month')
plt.legend()
plt.show()
```

### 5. Read face cream and facewash product sales data and show it using the bar chart.

```python
plt.figure(figsize=(16,7))
plt.bar(company_sales['month_number']-0.25,company_sales['facecream'],width=0.25,label='Face Cream Sales Data',align='edge')
plt.bar(company_sales['month_number']+0.25,company_sales['facewash'],width=-0.25,label='Face Cream Sales Data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Face Cream and Face Wash Sales Data')
plt.title('Face Cream and Face Wash sales Data')
plt.legend()
plt.show()
```

### 6. Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
plt.figure(figsize=(16,7))
plt.bar(company_sales['month_number'],company_sales['bathingsoap'],label='Bathing Soap Sales Data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales Data')
plt.title('Bathing Soap Sales Data')
plt.savefig('Bathing Soap.png')
plt.legend()
plt.show()
```

### 7. Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.figure(figsize=(16,7))
plt.hist(company_sales['total_profit'],label='Total Profit Data')
plt.xlabel('Actual Profit')
plt.ylabel('Profit Range')
plt.title('Total Profit')
plt.legend()
plt.show()
```

### 8. Calculate total sale data for last year for each product and show it using a Pie chart.

```python
total_sales_data=[sum(company_sales['facecream']), sum(company_sales['facewash']), sum(company_sales['toothpaste']),
                  sum(company_sales['bathingsoap']), sum(company_sales['shampoo']), sum(company_sales['moisturizer'])]
labels = ('Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer')
plt.pie(total_sales_data, labels=labels, autopct='%1.1f%%')
plt.show()
```

### 9. Read Bathing soap facewash of all months and display it using the Subplot

```python
a, arr = plt.subplots(2, sharex=True)
arr[0].plot(company_sales['month_number'],company_sales['bathingsoap'],label='Bathing Soap Sales Data', color='k', marker='o', linewidth=3)
arr[0].set_title('Bathing Soap Sales Data')
arr[1].plot(company_sales['month_number'],company_sales['facewash'],label='Face Wash Sales Data', color='r', marker='o', linewidth=3)
arr[1].set_title('Face Wash Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sales Data')
plt.show()
```

### 10. Read all product sales data and show it using the stack plot.

```python
plt.figure(figsize=(16,7))
plt.plot([],[],color='m', label='Face Cream Sales Data', linewidth=5)
plt.plot([],[],color='c', label='Face Wash Sales Data', linewidth=5)
plt.plot([],[],color='r', label='Tooth Paste Sales Data', linewidth=5)
plt.plot([],[],color='k', label='Bathing Soap Sales Data', linewidth=5)
plt.plot([],[],color='g', label='Shampoo Sales Data', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer Sales Data', linewidth=5)
plt.stackplot(company_sales['month_number'],company_sales['facecream'],company_sales['facewash'],company_sales['toothpaste'],
              company_sales['bathingsoap'],company_sales['shampoo'],company_sales['moisturizer'], colors = ['m','c','r','k','g','y'])
plt.ylabel('All Product Sales Data')
plt.xlabel('Month Number')
plt.legend()
plt.show()
```

### 11. Get total profit of all months and show line plot with the following Style properties.
### 12. Generated line plot must include following Style properties:
    ##### Line Style dotted and Line-color should be red
    ##### Show legend at the lower right location.
    ##### X label name = Month Number
    ##### Y label name = Sold unit’s number
    ##### Add a circle marker.
    ##### Line marker color as red
    ##### Line width should be 3

```python
plt.plot(company_sales['month_number'],company_sales['total_profit'], label='Total Profit', color='r', marker='o', markerfacecolor='k',
         linestyle='--', linewidth=3)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel("Sold Unit's Number")
plt.show()
```
